```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab as pl
```

```python
file_path = "/Users/atenafazeli/Desktop/ML/None Linear Regression/China_gdp/china_gdp.csv"

df = pd.read_csv(file_path)

df.head()
```

#### Plotting Data

```python
plt.figure (figsize = (8,5))
x_data, y_data = (df["Year"].values, df["Value"].values)
plt.plot (x_data, y_data, 'ro')
plt.xlabel ('Year')
plt.ylabel ('GDP')
plt.show
```

### Choosing the model  ( y_ = 1 / 1 + e ** (B-1(x-b-2)))

```python
x = np.arange (- 6, 6, 0.1)
y = 1.0 / (1.0 + np.exp(-x))

plt.plot(x, y)
plt.xlabel ('Independent Variable')
plt.ylabel ('Dependent Variable')
plt.show()

```

#### Building the model

```python
def sigmoid (x, Beta_1, Beta_2):
    y = 1 / (1 + np.exp(-Beta_1 * (x - Beta_2)))
    return y
```

### Normalize Data

```python
xdata = x_data / max (x_data)
ydata = y_data / max (y_data)
```

#### curve_fit to fit sigmoid function
#### Calculate Beta_1 and Beta_2
#### popt are optimized parameters

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit (sigmoid, xdata, ydata)
print ("beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

#### Plot regression model 

```python
x = np. linspace (1960, 2015, 55)
x = x/ max(x)
plt.figure  (figsize=(8, 5))
y = sigmoid (x, *popt)
plt. plot(xdata, ydata, 'ro', label='data' )
plt.plot (x, y, linewidth=3.0, label= 'fit')
plt.legend(loc='best')
plt.ylabel ('GDP')
plt.xlabel ('Year')
plt.show()
```
